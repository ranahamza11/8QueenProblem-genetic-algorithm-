## 8 Queen problem using genetic algorithm

```python
import random
def generateList():
    l = []
    for i in range(8):
        l.append(int(random.random()*8))
    return list(l)  



def calculateFitness(l):
    kill = {}
    for ind, val in enumerate(l):

    #forward
        mid= val
        up = val-1
        low = val+1
        for ind1, val1 in enumerate(l):
            if ind1 > ind:
                if val1 == up:
                    kill[ind1] = 1
                    kill[ind] = 1
                elif val1 == mid:
                    kill[ind1] = 1
                    kill[ind] = 1
                elif val1 == low:
                    kill[ind1] = 1
                    kill[ind] = 1
                up -= 1
                low += 1

    #backward
        mid= val
        up = val-1
        low = val+1
        for ind1, val1 in reversed(list(enumerate(l))):
      
            if ind1 < ind:
                if val1 == up:
                    kill[ind1] = 1
                    kill[ind] = 1
                elif val1 == mid:
                    kill[ind1] = 1
                    kill[ind] = 1
                elif val1 == low:
                    kill[ind1] = 1
                    kill[ind] = 1
                up -= 1
                low += 1

    return (len(kill))


def chooseBestFitness(fitness, data):
    small = fitness[0]
    smallInd = 0
    for i, val in enumerate(fitness):
        if small > val:
            samll = val
            smallInd = i
    if smallInd+1 < 4:
        small = fitness[smallInd + 1]
        second_smallInd = smallInd + 1
    else:
        small = fitness[smallInd - 1]
        second_smallInd = smallInd - 1
        
    for i, val in enumerate(fitness):
        if i != smallInd:
            if small > val and data[smallInd] != data[i]:
                samll = val
                second_smallInd = i
    return (smallInd, second_smallInd)

def crossOver(data):
    slicePos = int(random.random()*8)
    data.append(data[0][0:slicePos] + data[1][slicePos:])
    data.append(data[1][0:slicePos]+ data[0][slicePos:])
    return data

def mutation(data):
    mutatePos = int(random.random()*8)
    mutateVal = int(random.random()*8)
    temp = list(data[0])
    temp[mutatePos] = mutateVal
    data.append(temp)
    mutatePos = int(random.random()*8)
    mutateVal = int(random.random()*8)
    temp = list(data[1])
    temp[mutatePos] = mutateVal
    data.append(temp)
    return data

#generate default data here
data = []
for i in range(4):
    data.append(generateList())
count = 0
while True:
    fitness = []
    for i, val in enumerate(data):
        fitness.append(calculateFitness(val))
    if 0 in fitness:
        break
    small, second_small = chooseBestFitness(fitness, data)
    data = [data[small], data[second_small]]
    data = crossOver(data)
    fitness = []
    for i, val in enumerate(data):
        fitness.append(calculateFitness(val))

    if 0 in fitness:
        break
    small, second_small = chooseBestFitness(fitness, data)
    data = [data[small], data[second_small]]
    data = mutation(data)
    
for i, val in enumerate(fitness):
    if val == 0:
        print(data[i])
```
